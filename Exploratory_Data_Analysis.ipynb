```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df_post_2020 = pd.read_csv('data/post_2020_data_clean.csv')
df_post_2020.head(10)
```

```python
df_post_2020.dtypes
```

```python
# Converting start/end time from object to datetime 

df_post_2020['Start_Time'] = pd.to_datetime(df_post_2020["Start_Time"])
df_post_2020['End_Time'] = pd.to_datetime(df_post_2020["End_Time"])
```

```python
ax = df_post_2020["Start_Time"].groupby(df_post_2020["Start_Time"].dt.date).count().plot(kind="bar")

# formatter = mdates.ConciseDateFormatter(mdates.AutoDateLocator(minticks=3, maxticks=7))

# ax.xaxis.set_major_locator(mdates.AutoDateLocator(minticks=3, maxticks=7))
# ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.show()
```

```python
ax = df_post_2020["Start_Time"].groupby(df_post_2020["Start_Time"].dt.hour).count().plot(kind="bar")

ax.xaxis.set_major_locator(mdates.HourLocator())
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
plt.show()
```
